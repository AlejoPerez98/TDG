```python
import os
import pandas as pd

# Ruta al archivo de Excel
file_path_cruce = r"C:\Users\nhenao\OneDrive - CAMARA DE COMERCIO DE CALI\Documentos\Comercio\Exportaciones\2025\Enero/Ene25.xlsx"

# Leer el archivo Excel
Cruce = pd.read_excel(file_path_cruce)

# Aplicar la lógica de la fórmula a la columna 'SUBPARTIDA'
def aplicar_formula(SUBPARTIDA):
    try:
        SUBPARTIDA_num = str(int(SUBPARTIDA))  # Convertir a número y luego a string para eliminar decimales
        if len(SUBPARTIDA_num) == 9:
            return '0' + SUBPARTIDA_num[:3]
        else:
            return SUBPARTIDA_num[:4]
    except ValueError:
        return pd.NA  # En caso de que la conversión falle, devolver NaN

# Crear la nueva columna '4_digitos'
Cruce['4_digitos'] = Cruce['SUBPARTIDA'].apply(aplicar_formula)

# Convertir la columna '4_digitos' a tipo numérico
Cruce['4_digitos'] = pd.to_numeric(Cruce['4_digitos'], errors='coerce')

# Imprimir las primeras filas del DataFrame con la nueva columna
print(Cruce[['SUBPARTIDA', '4_digitos']].head())

# Guardar el DataFrame actualizado en un nuevo archivo Excel
output_file_cruce = r"C:\Users\nhenao\OneDrive - CAMARA DE COMERCIO DE CALI\Documentos\Comercio\Exportaciones\2025\Enero/Ene25.xlsx"
Cruce.to_excel(output_file_cruce, index=False)

print(f"Archivo actualizado guardado como: {output_file_cruce}")
```
